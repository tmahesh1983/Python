## Styling and Labelling

```python
x = np.arange(-3, 3.1, 0.1)
y = x ** 2
plt.plot(x, y, c='g', linestyle='dashed', marker = 'o', 
         markersize = 5, markerfacecolor='red')
plt.title('Quadratic func')
plt.xlabel('x-values')
plt.ylabel('x^2')

plt.show()
```

```python
x = np.arange(-3, 3.1, 0.1)
y = x ** 2
plt.plot(x, y, c='g', linestyle='dashed', marker = 'o', 
         markersize = 5, markerfacecolor='red')
plt.title('Quadratic func')
plt.xlabel('x-values')
plt.ylabel('x^2')
plt.xlim(0, 3)

plt.show()
```

```python
x = np.arange(-3, 3.1, 0.1)

plt.plot(x, x**2, label = 'x^2')
plt.plot(x, x**3, label = 'x^3')
plt.title('Quadratic func')
plt.xlabel('x-values')
plt.ylabel('Y-Values')
plt.legend()

plt.show()
```

```python
x = np.arange(-3, 3.1, 0.1)

plt.plot(x, x**2, label = 'x^2')
plt.plot(x, x**3, label = 'x^3')
plt.title('Quadratic func')
plt.xlabel('x-values')
plt.ylabel('Y-Values')
plt.legend()
plt.text(x=0, y= 0, s='(0, 0)')
plt.text(x=0, y= 20, s='Hello world')

plt.show()
```

```python
x = np.arange(-3, 3.1, 0.1)

plt.plot(x, x**2, label = 'x^2')
plt.plot(x, x**3, label = 'x^3')
plt.title('Quadratic func')
plt.xlabel('x-values')
plt.ylabel('Y-Values')

plt.text(x=0, y= 0, s='(0, 0)')
plt.text(x=0, y= 20, s='Hello world')
plt.legend()
plt.grid(True)
# plt.axis('off')

plt.show()
```

## Subplot

```python
x = np.arange(1, 10)
y = x ** 2
```

```python
fig = plt.figure()
plt.subplot(231)
plt.plot(x, y)
plt.xlabel('x-val')

plt.subplot(233)
plt.scatter(x, y)

plt.subplot(235)
plt.scatter(x, -x**3)
plt.show()
```

## Another way of sub plots
#### Object oriented way

```python
f, ax = plt.subplots(3, 1)
ax[0].plot(x, y)
ax[1].plot(x, -y)
ax[2].plot(x, np.log(y))
```

```python
a = np.arange(-1, 1, 0.1)
b = a
X, Y = np.meshgrid(a, b)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(X, Y, X**2 + Y**2)
plt.show()
```
