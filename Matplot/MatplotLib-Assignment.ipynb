## Vertically stacked graph

```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd 
np.random.seed(0)
df = pd.DataFrame(data={'1':np.random.randint(0, 100, 10), '2':np.random.randint(0, 100, 10)})
plt.bar(df.index.values, df['1'])
plt.bar(df.index.values, df['2'], bottom = df['1']) #Complete the code

plt.legend(['1', '2'])
plt.show()
```

## Subplots

```python
import matplotlib.pyplot as plt
pi=3.14
x = np.linspace(0,2*pi, 100)

#a
ax1 = plt.subplot(2,1,1)    
plt.xlim(0, 2*pi)
plt.plot(x, np.sin(2*x))
plt.title('subplot(2,1,1)')

#b
ax2 = plt.subplot(2,1,2)
plt.xlim(0, 2*pi)
plt.plot(x, np.sin(10*x), 'g')
plt.title('subplot(2,1,2)')

#c
plt.axes(ax1)
plt.plot(x, np.cos(2*x), 'r--')

plt.show()
```
