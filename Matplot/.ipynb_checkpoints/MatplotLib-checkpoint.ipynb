```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_values = [0, 1, 2]
y_values = [3, 5, 10]
```

```python
plt.plot(x_values, y_values)
plt.show()
```

```python
# X value is optional, if we are not providing it will take as 0,1,2...
plt.plot(y_values)
plt.show()
```

```python
x = np.arange(-10, 10)
y = x ** 2
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(-2, 3)
y = x ** 2
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(-3, 3.1, 0.1)
y = x ** 2
plt.plot(x, y)
plt.xticks(np.arange(-3, 3.1, 0.3), rotation=90)
plt.yticks(np.arange(0, 8.1, 0.5))
plt.show()
```

## Styling and Labelling

```python
x = np.arange(-3, 3.1, 0.1)
y = x ** 2
plt.plot(x, y, c='g', linestyle='dashed', marker = 'o', 
         markersize = 5, markerfacecolor='red')
plt.title('Quadratic func')
plt.xlabel('x-values')
plt.ylabel('x^2')

plt.show()
```

```python
x = np.arange(-3, 3.1, 0.1)
y = x ** 2
plt.plot(x, y, c='g', linestyle='dashed', marker = 'o', 
         markersize = 5, markerfacecolor='red')
plt.title('Quadratic func')
plt.xlabel('x-values')
plt.ylabel('x^2')
plt.xlim(0, 3)

plt.show()
```

```python
x = np.arange(-3, 3.1, 0.1)

plt.plot(x, x**2, label = 'x^2')
plt.plot(x, x**3, label = 'x^3')
plt.title('Quadratic func')
plt.xlabel('x-values')
plt.ylabel('Y-Values')
plt.legend()

plt.show()
```

```python
x = np.arange(-3, 3.1, 0.1)

plt.plot(x, x**2, label = 'x^2')
plt.plot(x, x**3, label = 'x^3')
plt.title('Quadratic func')
plt.xlabel('x-values')
plt.ylabel('Y-Values')
plt.legend()
plt.text(x=0, y= 0, s='(0, 0)')
plt.text(x=0, y= 20, s='Hello world')

plt.show()
```

```python
x = np.arange(-3, 3.1, 0.1)

plt.plot(x, x**2, label = 'x^2')
plt.plot(x, x**3, label = 'x^3')
plt.title('Quadratic func')
plt.xlabel('x-values')
plt.ylabel('Y-Values')

plt.text(x=0, y= 0, s='(0, 0)')
plt.text(x=0, y= 20, s='Hello world')
plt.legend()
plt.grid(True)
# plt.axis('off')

plt.show()
```
