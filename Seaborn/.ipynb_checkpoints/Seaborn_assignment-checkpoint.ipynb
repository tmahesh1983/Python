### Male and female passengers

##### which observation is incorrect wrt to the gender of passengers and survival?

##### In the overall number of passengers who survived, the majority were females.

```python
titanic['sex'].value_counts()
```

##### The number of passengers in the age bracket [20,40] is higher than other age brackets.

```python
temp = titanic[titanic['survived'] == 1]
plt.hist(temp['age'], bins=4)
plt.show()
```

```python
plt.hist(titanic['age'], bins=4)
plt.show()
```

##### The number of young male passengers who survived is greater than the number of young male passengers who didn't survive.

```python
male_passanger = titanic[(titanic['survived'] == 1) & (titanic['sex'] == 'male')]
male_passanger
```

```python
sns.histplot(data=male_passanger, x='age', bins=4)
plt.show()
```
