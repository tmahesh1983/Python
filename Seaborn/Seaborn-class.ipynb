```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
iris['species'].value_counts()
```

```python
iris['petal_length']
```

## Histogram

```python
plt.hist(iris['petal_length'], bins=20)
plt.show()
```

```python
sns.histplot(data=iris, x='petal_length', bins=25)
plt.show()
```

## KDE
 ##### Kernel density estimation

```python
sns.kdeplot(data=iris, x='petal_length')
plt.show()
```

```python
#hue = always categorical column.
sns.kdeplot(data=iris, x='petal_length', hue = 'species')
plt.show()
```

```python
sns.histplot(data=iris, x='petal_length', bins=25, hue = 'species')
plt.show()
```

```python
sns.scatterplot(data=iris, x='petal_length', y='petal_width')
plt.show()
```

```python
sns.jointplot(data=iris, x = 'sepal_length', y='sepal_width', hue = 'species')
plt.show()
```

```python
sns.jointplot(data=iris, x = 'sepal_length', y='sepal_width', kind = 'hex')
plt.show()
```

```python
sns.pairplot(data=iris, hue='species')
plt.show()
```

## Categorical plot

```python
sns.countplot(data=iris, x='species')
plt.show()
```

```python
sns.countplot(data=iris, y='species')
plt.show()
```

```python
sns.boxplot(data=iris, y='petal_length')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris, x='species', y='sepal_length')
```

```python
sns.violinplot(data=iris, y='petal_length', x='species')
```

## Fiding the correlation

```python
corr = iris.corr()
```

```python
sns.heatmap(corr, cmap='winter')
```
