#### Mask email

```python
def mask_email(s):
    if '@' in s:
        name, domain = s.split('@')
        return f"{name[0]}xxxx{name[-1]}@{domain}"
```

```python
mask_email("tmahesh1983@gmail.com")
```

```python
mask_email("tmahesh1983.com")
```

```python
mask_email("tmahesh1983@gmail")
```

```python
import re #import regex library
def is_valid_email(s):
    email_pattern = "^\w+([\.-]?\w+)*@\w+([\.-]?\w+)*(\.\w{2,3})+$"
    res = re.search(email_pattern, s)
    print (res)
    
    if res:
        return True
    else:
        return False
```

```python
is_valid_email("tmahesh1983.com")
```

```python
is_valid_email('tmahesh1983@gmail')
```

```python
is_valid_email('_tmahesh1983@gmail.com')
```

## Regular Expression

```python
regex = "^\w+([\.-]?\w+)*@\w+([\.-]?\w+)*(\.\w{2,3})+$" # highly non readable format

# https://docs.python.org/2/library/re.html
regex_verbose = re.compile(r"""            # VERY readable and easy to understand
                        ^\w+([\.-]?\w+)*   # start, \w+
                        @                  # single @ sign
                        \w+([\.-]?\w+)*    # Domain name
                        (\.\w{2,3})+$      # .com, .ac.in
                        """, re.VERBOSE | re.IGNORECASE) 

res = regex_verbose.match("abcd@iisc.ac.in")
print(res.string)
print(res)
```
