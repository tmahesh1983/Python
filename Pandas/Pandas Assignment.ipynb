## Q12. preprocess the dataframe

```python
import pandas as pd
import numpy as np
```

```python
def preprocess(df):
    

  """
    Takes df as input, do the preprocessing as asked.
    Return the preprocessed dataframe
  """
  df.dropna(how='all', inplace=True)
  df['Roll_ID'] = df['Roll_ID'].fillna(0).astype('int')
  df['Name'] = df['Name'].fillna('Anonymous')
  df['Marks'] = df['Marks'].median()
    
  return df.T
```

```python
df1 = pd.DataFrame()
df1['Roll_ID'] = pd.Series(data = [412, np.nan, 456, np.nan, 434, 429, 418])
df1['Name'] = pd.Series(data = ['John', 'Mitra', 'Ritz', np.nan, 'Anny', 'Hema', np.nan])
df1['Marks'] = pd.Series(data = [np.nan, 32, 25, np.nan, 35, 28, 38])
df1
```

```python
df1.dropna(how='all', inplace=True)
df1
```

```python
preprocess(df1)
```

## Q16. Concatenation-average

```python
import pandas as pd 
import numpy as np
df1 = pd.DataFrame({"name": ['tobey', 'peter', 'chris', 'pratt'],'chem_score':[10, 9, 8, 7], 'phy_score':[7, 8, 9, 10], 'Hindi_score':[9, 9, 9, 9]})  
df2 = pd.DataFrame({"name": ['chris', 'pratt', 'andrew', 'tom'],'chem_score':[10, 10, 10, 9], 'maths_score':[6, 6, 7, 9], 'eng_score':[9, 10, 10, 9]})
chem_avg =pd.concat([df1['chem_score'], df2['chem_score']], axis = 0).mean()
df3 = pd.merge(df1, df2, how='inner', on=['name'])
df3.set_index('name', inplace=True)
df4 = df3.T
print(df4.mean())
array = np.array([chem_avg, [np.mean(df4['chris']), np.mean(df4['pratt'])]])
array
```
