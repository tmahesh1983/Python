## Student Filtering

```python
student_dic = {"Name":['Himanshu', 'Robert', 'Karie', 'Rohan', 'John'], 
               "Age": [15, 14, 15, 16, 17], "Percentage":[80, 77, 83, 45, 70], 
               "Stream":['Commerce', 'Science', 'Arts', 'Commerce', 'Science']}
student_df = pd.DataFrame(student_dic)
student_df
```

```python
import pandas as pd
def student_filter(df):
    '''
    input:
    df -> the dataframe provided to the function
    
    output:
    df_new -> the selected students of the dataframe
    '''
    df_new = None
    
    # Your code starts here
    df_new = df[(df['Percentage'] >= 75) & ((df["Stream"] == 'Commerce') | (df["Stream"] == 'Arts'))]['Name']

    # Your code ends here
    return df_new
    
```

```python
res_df = student_filter(student_df)
res_df
```

## Location/Divide

```python
location_dic = {"City\tState":['Kolkata\tWest Bengal', 'Chennai\tTamil Nadu', 'Hyderabad\tTelengana', 'Bangalore\tKarnataka']}
location_df = pd.DataFrame(location_dic)
location_df
```

```python
def location_divide(df):
    '''
    input:
    df -> the dataframe provided to the function
    
    output:
    df_new -> the new preprocessed dataframe
    '''
    df_new = None    
    # Your code starts here
    df_new_columns = df.columns[0].split()
    df_new_value = df[df.columns[0]].str.split("\\t")
    list1 = []
    list2 = []
    for i in range(len(df_new_value)):
        list1.append(df_new_value[i][0])
        list2.append(df_new_value[i][1])
    print(list1)
    print(list2)
    print(df_new_value[1])
    df_new = pd.DataFrame({df_new_columns[0] : list1, df_new_columns[1] : list2}) 
    # Your code ends here
    
    return df_new
```

```python
location_divide(location_df)
```
