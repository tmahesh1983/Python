```python
import pandas as pd
def solve(reg_id, reg_dates):
    """
    Complete the function to return a Pandas Dataframe object as per the description.
    """
    df=None
    # YOUR CODE GOES HERE
    df = pd.DataFrame({'RID':reg_id, 'RDate':reg_dates})
    df['RMonth'] = pd.to_datetime(df['RDate']).dt.month
    df['RMonth'] = df['RMonth'].astype(str).str.rjust(2, "0")      
    df['RYear'] = pd.to_datetime(df['RDate']).dt.year
    df['RDay'] = pd.to_datetime(df['RDate']).dt.day
    df['RDay'] = df['RDay'].astype(str).str.rjust(2, "0") 
    return df
```

```python
# rid = [498, 721, 375, 464, 813, 853]
# rdates = ['2015-12-06', '2011-12-27', '2015-09-07', '2012-12-21', '2020-02-13', '2015-06-09']

rid = [498, 721, 375, 464, 813, 853]
rdates = ['2015-12-06', '2011-12-27', '2015-09-07', '2012-12-21', '2020-02-13', '2015-06-09']
solve(rid, rdates)
```
